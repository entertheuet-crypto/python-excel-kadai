```python
import openpyxl
import pandas as pd
from openpyxl.styles import PatternFill
```

```python
df1 = pd.read_excel('2022_年間売上表.xlsx')
df2 = pd.read_excel('2023_年間売上表.xlsx')

result = pd.concat([df1, df2], ignore_index=True)

grouped = result.groupby(['商品', '売上年'], as_index=False).sum()

writer = pd.ExcelWriter('売上集計表.xlsx')
grouped.to_excel(writer, sheet_name='sheet1', index=False)
writer.close()

wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb.active

for row in ws['A1:C1']:
    for cell in row:
        cell.fill = PatternFill(patternType='solid', fgColor='F2F2F2')

wb.save('売上集計表.xlsx')
```
