```python
import openpyxl
import time
from datetime import datetime
```

```python
workbook = openpyxl.Workbook()
sheet = workbook.active

sheet['B2'].value = '請求書'
sheet.insert_rows(3)

sheet['B4'].value = '株式会社ABC'
sheet['F4'].value = 'No.'
sheet['G4'].value = '0001'

sheet['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['F5'].value = '日付'
sheet['G5'] = time.strftime('%Y/%m/%d')

sheet['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'

sheet['B7'].value = '担当者名:鈴木一郎 様'
sheet.insert_rows(8, 2)

start_row = 10
header = ['商品名', '数量', '単価', '金額']
for col, value in enumerate(header, start=2):
    sheet.cell(row=start_row, column=col, value=value)

data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000]
]
for row_idx, row_data in enumerate(data, start=start_row + 1):
    for col_idx, value in enumerate(row_data, start=2):
        sheet.cell(row=row_idx, column=col_idx, value=value)

sheet['E13'] = 35000
sheet.insert_rows(14, 1)

sheet['B15'].value = '小計'
sheet['E15'] = 35000
sheet['B16'].value = '消費税'
sheet['E16'] = 3500
sheet['B17'].value = '合計'
sheet['E17'] = 38500

today = datetime.today()
backup_filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

workbook.save(backup_filename)
```
